```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
ind_f = cv2.imread("india-flag.svg.webp", cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(ind_f, 'gray')
```

```python
ind_f.shape
```

```python
print("Number of integers required =", len(ind_f.flatten()))
```

```python
u,d,v_t = np.linalg.svd(ind_f)
```

```python
d_rs = np.zeros((u.shape[0], v_t.shape[0]))
d_rs[0:u.shape[0], 0:u.shape[0]] = np.diag(d)

pre_c = len(d) + len(u.flatten()) + len(v_t.flatten())
```

```python
img_fr = u @ d_rs @ v_t
```

```python
plt.imshow(img_fr, 'gray')
```

```python
new_u = np.array([])
new_v_t = np.array([])
new_d = np.array([])
for i in range(len(d)):
    
    if i < 10:
        new_u = np.append(new_u, u[i])
        new_v_t = np.append(new_v_t, v_t[:,i])
        new_d = np.append(new_d, d[i])
```

```python
components = 4
new_u = u[:,:components]
new_v_t = v_t[:components,:]
new_d = d[:components]
```

```python
new_d.shape
```

```python
print(new_u.shape, new_d.shape, new_v_t.shape)
```

```python
new_d_rs = np.zeros((new_u.shape[0], new_v_t.shape[0]))
new_d_rs = np.diag(new_d)

new_d_rs.shape
```

```python
img_fr_c = new_u @ new_d_rs @ new_v_t
```

```python
plt.imshow(img_fr_c, 'gray')
```
